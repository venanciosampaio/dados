```python
import matplotlib.pyplot as plt
import numpy as np

# 1. Dados Fictícios de Vendas
# Meses (eixo X)
meses = ['Jan', 'Fev', 'Mar', 'Abr', 'Mai', 'Jun', 'Jul', 'Ago', 'Set', 'Out', 'Nov', 'Dez']
# Valores de Vendas (em milhares, eixo Y)
vendas = [25, 30, 28, 35, 40, 45, 42, 50, 48, 55, 60, 65]

# 2. Configuração do Gráfico
plt.figure(figsize=(10, 6)) # Define o tamanho da figura (largura, altura em polegadas)

# 3. Criação do Gráfico de Barras
plt.bar(meses, vendas, color='skyblue')

# 4. Adição de Títulos e Rótulos
plt.title('Vendas Mensais (Dados Fictícios)', fontsize=16)
plt.xlabel('Mês', fontsize=12)
plt.ylabel('Vendas (em Milhares de R$)', fontsize=12)

# Opcional: Adicionar os valores exatos em cima das barras
for i, v in enumerate(vendas):
    plt.text(i, v + 1, str(v) + 'K', ha='center', va='bottom', fontsize=10)

# Opcional: Adicionar uma linha de grade para melhor leitura
plt.grid(axis='y', linestyle='--', alpha=0.7)

# 5. Exibir o Gráfico
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# --- 1. Geração de Dados Probabilísticos (Distribuição Normal) ---
# Simulação de vendas em 1000 dias.
numero_de_dias = 1000
# Média esperada de vendas diárias
media_vendas = 150
# Desvio Padrão (variabilidade) das vendas diárias
desvio_padrao = 20

# Gerar os dados de vendas usando a distribuição normal (Gaussiana)
# np.random.normal(local=média, scale=desvio_padrão, size=tamanho)
vendas_diarias = np.random.normal(loc=media_vendas, scale=desvio_padrao, size=numero_de_dias)

# Garantir que as vendas não sejam negativas (embora raro com a distribuição normal)
vendas_diarias = np.maximum(0, vendas_diarias).astype(int)

# --- 2. Configuração do Histograma (Gráfico Probabilístico) ---
plt.figure(figsize=(10, 6))

# Criação do Histograma
# bins=30 define 30 barras para agrupar os dados
# density=True normaliza a área total para 1, mostrando a Probabilidade
plt.hist(vendas_diarias, bins=30, color='teal', edgecolor='black', alpha=0.7, density=True)

# Opcional: Adicionar a Curva de Distribuição Normal teórica
# Isso ajuda a visualizar como os dados se ajustam à distribuição esperada
xmin, xmax = plt.xlim()
x = np.linspace(xmin, xmax, 100)
p = (1/(desvio_padrao * np.sqrt(2 * np.pi))) * np.exp(-((x - media_vendas)**2 / (2 * desvio_padrao**2)))
plt.plot(x, p, 'k', linewidth=2)
# Adicionar texto com os parâmetros da distribuição
title = f"Fit da Distribuição Normal: $\mu$ = {media_vendas:.2f}, $\sigma$ = {desvio_padrao:.2f}"
plt.title(title, fontsize=14)

# --- 3. Adição de Títulos e Rótulos ---
plt.title('Distribuição de Frequência de Vendas Diárias (1000 Dias)', fontsize=16)
plt.xlabel('Vendas Diárias (Unidades)', fontsize=12)
plt.ylabel('Densidade de Probabilidade', fontsize=12)
plt.grid(axis='y', linestyle='--', alpha=0.6)

# --- 4. Exibir o Gráfico ---
plt.show()
```
