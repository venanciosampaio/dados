```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.mixture import GaussianMixture


#Define uma semente aleatória para garantir a reprodutividade dos resultados
np.random.seed(42)

#Criacção de dois Conjuntos de dados  ficticos
data1 = np.random.normal(loc=-5, scale=1, size=300) # 300 ponto com média -5 e desvio padrão 1
data2 = np.random.normal(loc=5, scale=1, size=700) # 700 ponto com média 5 e desvio padrão 1

#Combina os dois conjuntos de Dados para formar um unico gráfico
data = np.concatenate([data1, data2])

#visualiza o conjunto de dados original com um gráfico
plt.hist(data, bins=50, alpha=0.5)
plt.title("Sucos Originais")
plt.xlabel("Valor")
plt.ylabel("Frequência")
plt.show()



#Cria e treina um modela de Mistura gaussiana com dois componentes  (dois picos)
gmm = GaussianMixture(n_components=2)
#Ajusta o modelo aos dados
gmm.fit(data.reshape(-1, 1))


#Usando o treinamento para gerar 1000 novos pontos de dados
samples = gmm.sample(1000)[0].flatten()

plt.hist(samples, bins=30, density=True, alpha=0.5, color = 'green', label='Suco Primavera')
plt.hist(data, bins=30, density=True, alpha=0.5, color = 'red', label='Sucos Originais')
plt.legend()
plt.title("Comparação: Suco Primavera x Suco Original")
plt.xlabel("Valor")

#Obtem as probabilidades de cada ponto pertencer a um dos dois componentes
#probabilities = gmm.predict_proba(data.reshape(-1, 1))
```
